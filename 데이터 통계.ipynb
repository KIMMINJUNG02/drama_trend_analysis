# 나의 완벽한 비서

```python
%%R
positive_counts1 <- apply(com1[, -c(1,2)], 2, function(col) {sum(col == "positive")})
positive_counts1
```

```python
%%R
negative_counts1 <- apply(com1[, -c(1,2)], 2, function(col) {sum(col == "negative")})
negative_counts1
```

```python
%%R
data1 <- data.frame(
  element = names(negative_counts1),
  positive = positive_counts1,
  negative = negative_counts1
)

data1_melted <- melt(data1, id.vars = "element", variable.name = "sentiment", value.name = "count")
print(data1_melted)
```

```python
%%R
# Stacked Bar Chart 생성
p1 <- ggplot(data1_melted, aes(x = element, y = count, fill = sentiment)) +
  geom_col() +
  scale_fill_manual(values = c("positive" = "blue", "negative" = "red")) +
  coord_flip() +
  geom_text(aes(label = count), position = position_stack(vjust = 0.5), size = 8, color = "white") +
  labs(title = "나의 완벽한 비서", x = "언급된 요소", y = "언급된 횟수") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 40, face = "bold"),
    axis.title.x = element_text(size = 20),                      # x축 제목 크기
    axis.title.y = element_text(size = 20),
    legend.title = element_text(size = 25),
    legend.text = element_text(size = 20),
    axis.text.y = element_text(size = 30),  # y축 항목 이름 크기 키우기
    axis.text.x = element_text(size = 20)  # x축 눈금 크기 조정
  )
ggsave("/content/drive/MyDrive/자기주도연구/언급횟수 그래프/나의 완벽한 비서graph.png", plot = p1, width = 12, height = 5, dpi = 300)
```

```python
%%R
print(paste("전체,", nrow(com1)))
(positive_counts1 + negative_counts1)*100/nrow(com1)
```
