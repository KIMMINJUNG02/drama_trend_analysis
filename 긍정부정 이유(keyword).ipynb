````python
from openai import OpenAI
import json
````

````python
# GPT-4o를 사용하여 댓글에서 인기 요소 및 감정 분석 함수
def analyze_comment(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    prompt = f"""
    다음은 드라마에 대한 댓글입니다:
    "{comment}"

    1. 댓글에서 언급된 드라마 인기 요소(연기, 캐스팅, 캐릭터, 스토리, 연출, 외모, OST, 장르, 대사, 사회적 의미, 편성/플랫폼, 제작진(작가/감독))를 추출하세요.
    2. 각 요소별로 긍정(positive)인지 부정(negative)인지 분석하세요.
    3. JSON 형식으로 반환하세요. 부가적인 설명 없이 JSON만 출력하세요. 예시:

    {{
        "연기": "positive",
        "스토리": "negative",
        "OST": "positive"
    }}
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "You are an AI that extracts key elements and sentiment from user comments."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.5
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_중증외상센터_after_filtering.csv')
````

````python
comment_list.drop(columns=["relation"], inplace=True)
````

````python
comment_list["연기"] = "N/A"
comment_list["캐스팅"] = "N/A"
comment_list["캐릭터"] = "N/A"
comment_list["스토리"] = "N/A"
comment_list["연출"] = "N/A"
comment_list["외모"] = "N/A"
comment_list["OST"] = "N/A"
comment_list["장르"] = "N/A"
comment_list["대사"] = "N/A"
comment_list["사회적 의미"] = "N/A"
comment_list["편성/플랫폼"] = "N/A"
comment_list["제작진(작가/감독)"] = "N/A"
````

````python
for i in range(len(comment_list)):
  key_output = analyze_comment(comment_list.loc[i, "comment"])
  key_output = key_output.strip("```json").strip("```").strip()
  key_output = json.loads(key_output)
  for key,value in key_output.items():
    comment_list.loc[i, key] = value
````

````python
comment_list.to_csv('comments_중증외상센터_keyword.csv', index=False)
````
