# 연기 (negative)

````python
# GPT-4o를 사용하여 댓글에서 인기 요소 및 감정 분석 함수
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT API를 사용하여 댓글에서 부정적인 이유를 문장 단위로 추출
    """
    prompt = f"""
    다음 댓글에서 연기에 대한 부정적인 이유를 문장 단위로 추출하세요.
    형식: [{{"actor": "배우 이름 또는 배우 그룹", "acting_element": "연기 요소", "reason": "부정적인 이유"}}, ...]

    - 배우가 언급되었지만 이유가 없을 경우, "reason"을 "NA"로 설정하세요.
    - 연기 요소(감정 표현, 눈빛 연기, 대사 전달력, 발음 등)만 추출하고, 다른 요소(캐스팅, 캐릭터, 스토리, 연출, 외모, OST, 장르, 대사, 사회적 의미(메시지), 편성/플랫폼, 제작진(작가, 감독))는 제외하세요.
    - 연기와 관련된 요소(감정 표현, 눈빛 연기, 대사 전달력 등)를 명확히 구분하여 제공하세요.
    - 구체적인 이유가 언급되었지만 배우가 없을 경우, 배우 이름이 없을 경우 "actor" 값을 NA로 설정하세요.
    - 어떤 경우에도 빈 리스트([])가 반환되지 않도록 하세요. 적어도 하나의 결과를 제공하세요. 위 조건들을 반영하되 언급된 요소들은 리스트에 반영하세요.

    댓글: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "너는 드라마 연기 분석 전문가야."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_오징어게임2_keyword.csv')
````

````python
acting_negative = comment_list[comment_list['연기'] == 'negative']
acting_negative = acting_negative.reset_index(drop=True)
acting_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(len(acting_negative)):
  print(i)
  negative_reason = detail_info(acting_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": acting_negative.loc[i, "comment"],
            "actor": reason["actor"],
            "acting_element": reason["acting_element"],
            "reason": reason["reason"]
        })
````

````python
# 유사 or 동일한 카테고리는 하나로 묶기
def clear_table(col1, col2):
  global grouped_df
  plus_comment = grouped_df.loc[col1,"reason"]+grouped_df.loc[col2,"reason"]
  grouped_df.at[col1,"reason"] = plus_comment
  plus_count = grouped_df.loc[col1,"count"]+grouped_df.loc[col2,"count"]
  grouped_df.at[col1,"count"] = plus_count
````

## 연기 세부 요소별

````python
df=pd.read_csv("오징어게임2_acting_negative.csv")
````

````python
grouped_df = df.groupby("acting_element")["reason"].apply(list).reset_index()
grouped_df["count"]=0
````

````python
for i in range(len(grouped_df)):
  grouped_df.loc[i,"count"]=len(df[df["acting_element"]==grouped_df.loc[i,"acting_element"]])
````

````python
for i in range(15,21):
  clear_table(21,i)
  grouped_df = grouped_df.drop(index=i)
````
