# 연기 (positive)

````python
# GPT-4o를 사용하여 댓글에서 인기 요소 및 감정 분석 함수
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT API를 사용하여 댓글에서 긍정적인 이유를 문장 단위로 추출
    """
    prompt = f"""
    다음 댓글에서 연기에 대한 긍정적인 이유를 문장 단위로 추출하세요.
    형식: [{{"actor": "배우 이름 또는 배우 그룹", "acting_element": "연기 요소", "reason": "긍정적인 이유"}}, ...]

    - 배우가 언급되었지만 이유가 없을 경우, "reason"을 "NA"로 설정하세요.
    - 연기 요소(감정 표현, 눈빛 연기, 대사 전달력, 발음 등)만 추출하고, 다른 요소(캐스팅, 캐릭터, 스토리, 연출, 외모, OST, 장르, 대사, 사회적 의미(메시지), 편성/플랫폼, 제작진(작가, 감독))는 제외하세요.
    - 연기와 관련된 요소(감정 표현, 눈빛 연기, 대사 전달력 등)를 명확히 구분하여 제공하세요.
    - 구체적인 이유가 언급되었지만 배우가 없을 경우, 배우 이름이 없을 경우 "actor" 값을 NA로 설정하세요.
    - 어떤 경우에도 빈 리스트([])가 반환되지 않도록 하세요. 적어도 하나의 결과를 제공하세요. 위 조건들을 반영하되 언급된 요소들은 리스트에 반영하세요.

    댓글: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "너는 드라마 연기 분석 전문가야."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
acting_positive = comment_list[comment_list['연기'] == 'positive']
acting_positive = acting_positive.reset_index(drop=True)
acting_positive
````

````python
comment9="매우 좋아하는 한지민과 이준혁 두 배우, 비주얼이 뛰어나고 분위기가 비슷하며 케미가 정말 좋다. 두 사람 모두 거만하거나 허세를 부리지 않고, 진지하게 연기에 몰입하는 배우들이다. 한지민은 이미 내 마음속 최고의 한국 여배우이며, 그녀가 출연하는 작품은 반드시 볼 것이다. 이준혁도 크게 성공하길 바라며, 앞으로 더 많은 로맨스 작품에 출연하길 기대한다. 이 정도의 비주얼을 활용하지 않는 것은 너무 아깝다. 진지하고 노력하는 배우이기에 더 많은 사람들이 주목하고 좋아해 주었으면 한다."
````

````python
positive_reason = detail_info(comment9)
positive_reason = positive_reason.strip("```json").strip("```").strip()
positive_reason = json.loads(positive_reason)
````

````python
for i in range(len(acting_positive)):
  print(i)
  positive_reason = detail_info(acting_positive.loc[i, "comment"])
  positive_reason = positive_reason.strip("```json").strip("```").strip()
  positive_reason = json.loads(positive_reason)
  for reason in positive_reason:
        results.append({
            "comment": acting_positive.loc[i, "comment"],
            "actor": reason["actor"],
            "acting_element": reason["acting_element"],
            "reason": reason["reason"]
        })
````

## 연기 세부 요소별 count

````python
for i in range(len(grouped_df)):
  grouped_df.loc[i,"count"]=len(df[df["acting_element"]==grouped_df.loc[i,"acting_element"]])
````
