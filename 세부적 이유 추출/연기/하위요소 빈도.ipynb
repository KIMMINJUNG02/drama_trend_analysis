```python
import pandas as pd
```

# 연기

```python
detail_negative = pd.read_csv("/content/drive/MyDrive/자기주도연구/세부적 이유 추출/연기/all_negative_reason.csv")
detail_positive = pd.read_csv("/content/drive/MyDrive/자기주도연구/세부적 이유 추출/연기/all_positive_reason.csv")
```

```python
detail = pd.concat([detail_negative, detail_positive], ignore_index=True)
```

```python
detail
```

```python
detail_count = detail.groupby("acting_element", as_index=False)["count"].sum()
detail_count
```

```python
detail_count.to_csv("/content/drive/MyDrive/자기주도연구/세부적 이유 추출/연기/acting_count.csv",index=False)
```
