````python
# GPT-4oë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ì¸ê¸° ìš”ì†Œ ë° ê°ì • ë¶„ì„ í•¨ìˆ˜
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT APIë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œ
    """
    prompt = f"""
    ë‹¤ìŒ ëŒ“ê¸€ì—ì„œ ìŠ¤í† ë¦¬ì— ëŒ€í•œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œí•˜ì„¸ìš”.
    í˜•ì‹: [{{"story_element": "ìŠ¤í† ë¦¬ ìš”ì†Œ", "reason": "ë¶€ì •ì ì¸ ì´ìœ "}}, ...]

    - ìŠ¤í† ë¦¬ ìš”ì†Œ(ì£¼ì œ, ì „ê°œ ë°©ì‹, í”Œë¡¯ êµ¬ì¡°, ì„œì‚¬ ëª°ì…ë„, ê°ì •ì  ë°˜ì‘ ë“±)ë§Œ ì¶”ì¶œí•˜ê³ , ë‹¤ë¥¸ ìš”ì†Œ(ìºìŠ¤íŒ…, ìºë¦­í„°, ì—°ê¸°, ì—°ì¶œ, ì™¸ëª¨, OST, ì¥ë¥´, ëŒ€ì‚¬, ì‚¬íšŒì  ì˜ë¯¸(ë©”ì‹œì§€), í¸ì„±/í”Œë«í¼, ì œì‘ì§„(ì‘ê°€, ê°ë…))ëŠ” ì œì™¸í•˜ì„¸ìš”.
    - ìŠ¤í† ë¦¬ì™€ ê´€ë ¨ëœ ìš”ì†Œ(ì£¼ì œ, ì „ê°œ ë°©ì‹, í”Œë¡¯ êµ¬ì¡°, ì„œì‚¬ ëª°ì…ë„, ê°ì •ì  ë°˜ì‘ ë“±)ë¥¼ ëª…í™•íˆ êµ¬ë¶„í•˜ì—¬ ì œê³µí•˜ì„¸ìš”.

    ëŒ“ê¸€: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "ë„ˆëŠ” ë“œë¼ë§ˆ ìŠ¤í† ë¦¬ ë¶„ì„ ì „ë¬¸ê°€ì•¼."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_ì¤‘ì¦ì™¸ìƒì„¼í„°_keyword.csv')
````

````python
story_negative = comment_list[comment_list['ìŠ¤í† ë¦¬'] == 'negative']
story_negative = story_negative.reset_index(drop=True)
story_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(len(story_negative)):
  print(i)
  negative_reason = detail_info(story_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": story_negative.loc[i, "comment"],
            "story_element": reason["story_element"],
            "reason": reason["reason"]
        })
````
