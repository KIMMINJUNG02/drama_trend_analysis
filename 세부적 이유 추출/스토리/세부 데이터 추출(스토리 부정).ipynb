```python
com1 = pd.read_csv('세부적 이유 추출/스토리/나의 완벽한 비서/negative/나의 완벽한 비서_story_negative_reason.csv')
com2 = pd.read_csv('세부적 이유 추출/스토리/모텔 캘리포니아/negative/모텔캘리포니아_story_negative_reason.csv')
com3 = pd.read_csv('세부적 이유 추출/스토리/오징어게임2/negative/오징어게임2_story_negative_reason.csv')
com4 = pd.read_csv('세부적 이유 추출/스토리/옥씨부인전/negative/옥씨부인전_story_negative_reason.csv')
com5 = pd.read_csv('세부적 이유 추출/스토리/원경/negative/원경_story_negative_reason.csv')
com6 = pd.read_csv('세부적 이유 추출/스토리/중증외상센터/negative/중증외상센터_story_negative_reason.csv')
```

```python
comment_list = pd.concat([com1, com2, com3, com4, com5, com6], ignore_index=True)
```

```python
# 유사 or 동일한 카테고리는 하나로 묶기
def clear_actor(col1, col2):
  global comment_list
  plus_comment = comment_list.loc[col1,"reason"]+comment_list.loc[col2,"reason"]
  comment_list.at[col1,"reason"] = plus_comment
  plus_count = comment_list.loc[col1,"count"]+comment_list.loc[col2,"count"]
  comment_list.at[col1,"count"] = plus_count
```

```python
def generate_summary(trend_data):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """GPT를 사용하여 스토리 요소에 대한 트렌드 요약 생성"""
    prompt = f"""
    다음은 시청자들의 반응을 포함하는 데이터셋입니다. 이 데이터를 기반으로 K-드라마에서 사람들이 어떤 부분을 싫어하고, 부정적인 반응을 보이는지 트렌드를 분석하세요.

    {trend_data.to_string(index=False)}

    분석할 내용:
    1. 시청자들이 싫어하는 부분
    2. 전반적인 감상 트렌드 (예: 스토리 흐름, 플롯 구조, 전개 방식 등)
    3. 앞으로 드라마 시장에서 개선되어야하는 방향

    형식:
    - **주요 트렌드 요약**: (트렌드 내용)
    - **시청자들이 싫어하는 요소**: (예시)
    - **기타 주목할 만한 사항**: (예시)
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "너는 드라마 스토리 분석 전문가야."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content.strip()
```

```python
def generate_final_summary(trend_data):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """GPT를 사용하여 스토리 요소에 대한 트렌드 요약 생성"""

    summaries_text = "\n\n".join(trend_data)

    prompt = f"""
    다음은 시청자들의 반응을 포함하는 데이터셋입니다. 이 데이터를 기반으로 K-드라마에서 사람들이 어떤 부분을 싫어하고, 부정적인 반응을 보이는지 트렌드를 분석하세요.

    {summaries_text}

    분석할 내용:
    1. 시청자들이 싫어하는 부분
    2. 전반적인 감상 트렌드 (예: 스토리 흐름, 플롯 구조, 전개 방식 등)
    3. 앞으로 드라마 시장에서 개선되어야하는 방향

    형식:
    - **주요 트렌드 요약**: (트렌드 내용)
    - **시청자들이 싫어하는 요소**: (예시)
    - **기타 주목할 만한 사항**: (예시)
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "너는 드라마 스토리 분석 전문가야."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content.strip()
```

```python
final_summary = generate_final_summary(trend_summary)
```
