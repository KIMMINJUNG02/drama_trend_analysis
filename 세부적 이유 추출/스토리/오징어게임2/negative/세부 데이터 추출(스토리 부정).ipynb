# 스토리 (positive)

````python
# GPT-4o를 사용하여 댓글에서 인기 요소 및 감정 분석 함수
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT API를 사용하여 댓글에서 부정적인 이유를 문장 단위로 추출
    """
    prompt = f"""
    다음 댓글에서 스토리에 대한 부정적인 이유를 문장 단위로 추출하세요.
    형식: [{{"story_element": "스토리 요소", "reason": "부정적인 이유"}}, ...]

    - 스토리 요소(주제, 전개 방식, 플롯 구조, 서사 몰입도, 감정적 반응 등)만 추출하고, 다른 요소(캐스팅, 캐릭터, 연기, 연출, 외모, OST, 장르, 대사, 사회적 의미(메시지), 편성/플랫폼, 제작진(작가, 감독))는 제외하세요.
    - 스토리와 관련된 요소(주제, 전개 방식, 플롯 구조, 서사 몰입도, 감정적 반응 등)를 명확히 구분하여 제공하세요.

    댓글: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "너는 드라마 스토리 분석 전문가야."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_오징어게임2_keyword.csv')
````

````python
comment_list
````

````python
story_negative = comment_list[comment_list['스토리'] == 'negative']
story_negative = story_negative.reset_index(drop=True)
story_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(827,len(story_negative)):
  print(i)
  negative_reason = detail_info(story_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": story_negative.loc[i, "comment"],
            "story_element": reason["story_element"],
            "reason": reason["reason"]
        })
````

````python
df = pd.DataFrame(results)
df
````

````python
df.to_csv("오징어게임2_story_negative.csv", index=False)
````

## 스토리 세부 요소별 count

````python
df=pd.read_csv("오징어게임2_story_negative.csv")
````

````python
grouped_df = df.groupby("story_element")["reason"].apply(list).reset_index()
grouped_df["count"]=0
````

````python
for i in range(len(grouped_df)):
  grouped_df.loc[i,"count"]=len(df[df["story_element"]==grouped_df.loc[i,"story_element"]])
````

````python
grouped_df
````

````python
# 유사 or 동일한 카테고리는 하나로 묶기
def clear_actor(col1, col2):
  global grouped_df
  plus_comment = grouped_df.loc[col1,"reason"]+grouped_df.loc[col2,"reason"]
  grouped_df.at[col1,"reason"] = plus_comment
  plus_count = grouped_df.loc[col1,"count"]+grouped_df.loc[col2,"count"]
  grouped_df.at[col1,"count"] = plus_count
````

````python
ae=[0]
````

````python
for i in ae:
  clear_actor(10,i)
  comment_list = comment_list.drop(index=i)
  comment_list
````

````python
grouped_df.to_csv("오징어게임2_story_negative_reason.csv", index=False)
````
