# ìºë¦­í„° (negative)

````python
# GPT-4oë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ì¸ê¸° ìš”ì†Œ ë° ê°ì • ë¶„ì„ í•¨ìˆ˜
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT APIë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œ
    """
    prompt = f"""
    ë‹¤ìŒ ëŒ“ê¸€ì—ì„œ ìºë¦­í„°ì— ëŒ€í•œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œí•˜ì„¸ìš”.
    í˜•ì‹: [{{"character": "ë°°ìš° ì´ë¦„ ë˜ëŠ” ìºë¦­í„° ì´ë¦„", "character_element": "ìºë¦­í„° ìš”ì†Œ", "reason": "ë¶€ì •ì ì¸ ì´ìœ "}}, ...]

    - ìºë¦­í„° ê´€ë ¨ ìš”ì†Œ(ì„±ê²©, ë§¤ë ¥ë„, í™”ë²•, ì„±ì¥/ë³€í™”, ì„œì‚¬, ê°œì—°ì„±, ì¼€ë¯¸, ì…ì²´ê°, ì½”ë¯¸ë”” ìš”ì†Œ, ì•…ì—­ ë§¤ë ¥, ë¹„ì£¼ì–¼ê³¼ ì‹±í¬ë¡œìœ¨ ë“±)**ì„ êµ¬ì²´ì ìœ¼ë¡œ ë¶„ë¥˜í•˜ì„¸ìš”.
    - ìºë¦­í„°ë‚˜ ë°°ìš°ê°€ ì–¸ê¸‰ë˜ì—ˆì§€ë§Œ ì´ìœ ê°€ ì—†ì„ ê²½ìš°, "reason"ì„ "NA"ë¡œ ì„¤ì •í•˜ì„¸ìš”.
    - ìºë¯¸ìŠ¤íŠ¸ë¦¬(ì¼€ë¯¸) ìš”ì†Œê°€ ì–¸ê¸‰ëœ ê²½ìš°, ì–´ëŠ ìºë¦­í„°ì™€ ì¼€ë¯¸ê°€ ë‚˜ìœì§€ ë°˜ë“œì‹œ í¬í•¨í•˜ì„¸ìš”.
      - ex: "character": "í•œì§€ë¯¼ & ì´ì¤€í˜", "character_element": "ì¼€ë¯¸", "reason": "ì„œë¡œ í‹°ê²©íƒœê²©í•˜ëŠ” ì¥ë©´ì´ ë„ˆë¬´ ë‹µë‹µí•˜ë‹¤."
    - ë‹¨ìˆœí•œ í‘œí˜„("ì‹«ìŒ", "ì•ˆ ì–´ìš¸ë¦¼")ì´ ì•„ë‹Œ, êµ¬ì²´ì ì¸ ì´ìœ (ì˜ˆ: ì¼€ë¯¸ê°€ ì˜ ì•ˆ ë§ëŠ”ê±° ê°™ë‹¤, ê°ì´ˆ ì—­í• ì´ ì–´ìƒ‰í•˜ë‹¤)ë¥¼ í¬í•¨í•˜ì„¸ìš”.
    - ìºë¦­í„° ìš”ì†Œ ì™¸ì˜ ë‹¤ë¥¸ ìš”ì†Œ(ìºìŠ¤íŒ…, ì—°ê¸°, ìŠ¤í† ë¦¬, ì—°ì¶œ, ì™¸ëª¨, OST, ì¥ë¥´, ì‚¬íšŒì  ì˜ë¯¸, ì œì‘ì§„)ëŠ” ì œì™¸í•˜ì„¸ìš”.
    - ë°°ìš° ì´ë¦„ì´ ì—†ì´ ìºë¦­í„°ë§Œ ì–¸ê¸‰ëœ ê²½ìš°ì—ë„ ì €ì¥í•˜ì„¸ìš”.**
    - ë°°ìš° ì´ë¦„ê³¼ ìºë¦­í„°ê°€ ë‘˜ ë‹¤ ì–¸ê¸‰ë˜ì§€ ì•Šê³  ìºë¦­í„° ìš”ì†Œë§Œ í‰ê°€ëœ ê²½ìš°, "character" ê°’ì„ "NA"ë¡œ ì„¤ì •í•˜ì„¸ìš”.**

    ëŒ“ê¸€: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "ë„ˆëŠ” ë“œë¼ë§ˆ ìºë¦­í„° ë¶„ì„ ì „ë¬¸ê°€ì•¼."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_ì¤‘ì¦ì™¸ìƒì„¼í„°_keyword.csv')
````

````python
comment_list
````

````python
character_negative = comment_list[comment_list['ìºë¦­í„°'] == 'negative']
character_negative = character_negative.reset_index(drop=True)
character_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(len(character_negative)):
  print(i)
  negative_reason = detail_info(character_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": character_negative.loc[i, "comment"],
            "character": reason["character"],
            "character_element": reason["character_element"],
            "reason": reason["reason"]
        })
````

````python
df = pd.DataFrame(results)
df
````

````python
df.to_csv("ì¤‘ì¦ì™¸ìƒì„¼í„°_character_negative.csv", index=False)
````

## ìºë¦­í„° ì„¸ë¶€ ìš”ì†Œë³„ count

````python
df=pd.read_csv("ì¤‘ì¦ì™¸ìƒì„¼í„°_character_negative.csv")
````

````python
grouped_df = df.groupby("character_element")["reason"].apply(list).reset_index()
grouped_df["count"]=0
````

````python
for i in range(len(grouped_df)):
  grouped_df.loc[i,"count"]=len(df[df["character_element"]==grouped_df.loc[i,"character_element"]])
````

````python
# ìœ ì‚¬ or ë™ì¼í•œ ì¹´í…Œê³ ë¦¬ëŠ” í•˜ë‚˜ë¡œ ë¬¶ê¸°
def clear_actor(col1, col2):
  global grouped_df
  plus_comment = grouped_df.loc[col1,"reason"]+grouped_df.loc[col2,"reason"]
  grouped_df.at[col1,"reason"] = plus_comment
  plus_count = grouped_df.loc[col1,"count"]+grouped_df.loc[col2,"count"]
  grouped_df.at[col1,"count"] = plus_count
````

````python
jm=[3]
````

````python
for i in jm:
  clear_actor(6,i)
  grouped_df = grouped_df.drop(index=i)
  grouped_df
````
