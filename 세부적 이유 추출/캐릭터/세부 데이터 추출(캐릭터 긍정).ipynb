```python
# 유사 or 동일한 카테고리는 하나로 묶기
def clear_table(col1, col2):
  global comment_list
  plus_comment = comment_list.loc[col1,"reason"]+comment_list.loc[col2,"reason"]
  comment_list.at[col1,"reason"] = plus_comment
  plus_count = comment_list.loc[col1,"count"]+comment_list.loc[col2,"count"]
  comment_list.at[col1,"count"] = plus_count
```

## 분석

```python
def generate_summary(trend_data):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """GPT를 사용하여 캐릭터 요소별 트렌드 요약 생성"""
    prompt = f"""
    다음은 시청자들의 반응을 포함하는 데이터셋입니다. 이 데이터를 기반으로 K-드라마에서 사람들이 어떤 부분을 선호하고, 긍정적인 반응을 보이는지 트렌드를 분석하세요.

    {trend_data.to_string(index=False)}

    분석할 내용:
    1. 시청자들이 선호하는 부분
    2. 전반적인 감상 트렌드 (예: 몰입감, 악역매력, 입체감, 임펙트 등)
    3. 앞으로 드라마 시장에서 계속 반영되었으면 하는 부분

    형식:
    - **주요 트렌드 요약**: (트렌드 내용)
    - **시청자들이 선호하는 요소**: (예시)
    - **기타 주목할 만한 사항**: (예시)
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "너는 드라마 캐릭터 분석 전문가야."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content.strip()
```

```python
def generate_final_summary(trend_data):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """GPT를 사용하여 캐릭터 요소에 대한 트렌드 요약 생성"""

    summaries_text = "\n\n".join(trend_data)

    prompt = f"""
    다음은 시청자들의 반응을 포함하는 데이터셋을 바탕으로 분석한 결과이다. 이 결과들을 바탕으로 K-드라마에서 사람들이 어떤 부분을 선호하고, 긍정적인 반응을 보이는지 트렌드를 분석하세요.

    {summaries_text}

    분석할 내용:
    1. 시청자들이 선호하는 부분
    2. 전반적인 감상 트렌드 (예: 몰입감, 악역매력, 입체감, 임펙트 등)
    3. 앞으로 드라마 시장에서 계속 반영되었으면 하는 부분

    형식:
    - **주요 트렌드 요약**: (트렌드 내용)
    - **시청자들이 선호하는 요소**: (예시)
    - **기타 주목할 만한 사항**: (예시)
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "너는 드라마 캐릭터 분석 전문가야."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )

    return response.choices[0].message.content.strip()
```

```python
final_summary_trend = generate_final_summary(trend_summary)
```

```python
with open("character_positive_trends_summary.md", "w", encoding="utf-8") as f: #appned : w->a
    f.write("드라마 최신 트랜드 분석(캐릭터 - positive)")  # 기존 내용과 구분
    f.write(final_summary_trend)

print("트렌드 요약이 character_positive_trends_summary.md 파일에 추가되었습니다.")
```
