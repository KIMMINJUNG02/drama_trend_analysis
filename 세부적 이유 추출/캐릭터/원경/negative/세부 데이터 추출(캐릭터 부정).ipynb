# 캐스팅 (negative)

````python
# GPT-4o를 사용하여 댓글에서 인기 요소 및 감정 분석 함수
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT API를 사용하여 댓글에서 부정적인 이유를 문장 단위로 추출
    """
    prompt = f"""
    다음 댓글에서 캐릭터에 대한 부정적인 이유를 문장 단위로 추출하세요.
    형식: [{{"character": "배우 이름 또는 캐릭터 이름", "character_element": "캐릭터 요소", "reason": "부정적인 이유"}}, ...]

    - 캐릭터 관련 요소(성격, 매력도, 화법, 성장/변화, 서사, 개연성, 케미, 입체감, 코미디 요소, 악역 매력, 비주얼과 싱크로율 등)**을 구체적으로 분류하세요.
    - 캐릭터나 배우가 언급되었지만 이유가 없을 경우, "reason"을 "NA"로 설정하세요.
    - 캐미스트리(케미) 요소가 언급된 경우, 어느 캐릭터와 케미가 나쁜지 반드시 포함하세요.
      - ex: "character": "한지민 & 이준혁", "character_element": "케미", "reason": "서로 티격태격하는 장면이 너무 답답하다."
    - 단순한 표현("싫음", "안 어울림")이 아닌, 구체적인 이유(예: 케미가 잘 안 맞는거 같다, 감초 역할이 어색하다)를 포함하세요.
    - 캐릭터 요소 외의 다른 요소(캐스팅, 연기, 스토리, 연출, 외모, OST, 장르, 사회적 의미, 제작진)는 제외하세요.
    - 배우 이름이 없이 캐릭터만 언급된 경우에도 저장하세요.**
    - 배우 이름과 캐릭터가 둘 다 언급되지 않고 캐릭터 요소만 평가된 경우, "character" 값을 "NA"로 설정하세요.**

    댓글: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "너는 드라마 캐릭터 분석 전문가야."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_원경_keyword.csv')
````

````python
comment_list
````

````python
character_negative = comment_list[comment_list['캐릭터'] == 'negative']
character_negative = character_negative.reset_index(drop=True)
character_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(len(character_negative)):
  print(i)
  negative_reason = detail_info(character_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": character_negative.loc[i, "comment"],
            "character": reason["character"],
            "character_element": reason["character_element"],
            "reason": reason["reason"]
        })
````

````python
df = pd.DataFrame(results)
df
````

````python
df.to_csv("원경_character_negative.csv", index=False)
````

## 캐릭터 세부 요소별 count

````python
df=pd.read_csv("원경_character_negative.csv")
````

````python
grouped_df = df.groupby("character_element")["reason"].apply(list).reset_index()
grouped_df["count"]=0
````

````python
for i in range(len(grouped_df)):
  grouped_df.loc[i,"count"]=len(df[df["character_element"]==grouped_df.loc[i,"character_element"]])
````

````python
# 유사 or 동일한 카테고리는 하나로 묶기
def clear_actor(col1, col2):
  global grouped_df
  plus_comment = grouped_df.loc[col1,"reason"]+grouped_df.loc[col2,"reason"]
  grouped_df.at[col1,"reason"] = plus_comment
  plus_count = grouped_df.loc[col1,"count"]+grouped_df.loc[col2,"count"]
  grouped_df.at[col1,"count"] = plus_count
````

````python
jm=[3]
````

````python
for i in jm:
  clear_actor(6,i)
  grouped_df = grouped_df.drop(index=i)
  grouped_df
````
