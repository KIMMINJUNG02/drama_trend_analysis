# ìºìŠ¤íŒ… (negative)

````python
# GPT-4oë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ì¸ê¸° ìš”ì†Œ ë° ê°ì • ë¶„ì„ í•¨ìˆ˜
def detail_info(comment):
    client = OpenAI(
      api_key = "my_gpt_api_key"
    )

    """
    GPT APIë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ“ê¸€ì—ì„œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œ
    """
    prompt = f"""
    ë‹¤ìŒ ëŒ“ê¸€ì—ì„œ ìºìŠ¤íŒ…ì— ëŒ€í•œ ë¶€ì •ì ì¸ ì´ìœ ë¥¼ ë¬¸ì¥ ë‹¨ìœ„ë¡œ ì¶”ì¶œí•˜ì„¸ìš”.
    í˜•ì‹: [{{"actor": "ë°°ìš° ì´ë¦„ ë˜ëŠ” ë°°ìš° ê·¸ë£¹", "reason": "ë¶€ì •ì ì¸ ì´ìœ "}}, ...]

    - ë°°ìš°ê°€ ì–¸ê¸‰ë˜ì—ˆì§€ë§Œ ì´ìœ ê°€ ì—†ì„ ê²½ìš°, "reason"ì„ "NA"ë¡œ ì„¤ì •í•˜ì„¸ìš”.
    - ë‹¤ë¥¸ ìš”ì†Œ(ìºìŠ¤íŒ…, ìºë¦­í„°, ìŠ¤í† ë¦¬, ì—°ì¶œ, ì™¸ëª¨, OST, ì¥ë¥´, ëŒ€ì‚¬, ì‚¬íšŒì  ì˜ë¯¸(ë©”ì‹œì§€), í¸ì„±/í”Œë«í¼, ì œì‘ì§„(ì‘ê°€, ê°ë…))ëŠ” ì œì™¸í•˜ì„¸ìš”.
    - êµ¬ì²´ì ì¸ ì´ìœ ê°€ ì–¸ê¸‰ë˜ì—ˆì§€ë§Œ ë°°ìš°ê°€ ì—†ì„ ê²½ìš°, ë°°ìš° ì´ë¦„ì´ ì—†ì„ ê²½ìš° "actor" ê°’ì„ NAë¡œ ì„¤ì •í•˜ì„¸ìš”.
    - ìœ„ ì¡°ê±´ë“¤ì„ ë°˜ì˜í•˜ë˜ ì–¸ê¸‰ëœ ìš”ì†Œë“¤ì€ ë¦¬ìŠ¤íŠ¸ì— ë°˜ì˜í•˜ì„¸ìš”.

    ëŒ“ê¸€: "{comment}"
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "system", "content": "ë„ˆëŠ” ë“œë¼ë§ˆ ìºìŠ¤íŒ… ë¶„ì„ ì „ë¬¸ê°€ì•¼."},
                  {"role": "user", "content": prompt}],
        temperature=0.7
    )
    print("response:")
    print(response)
    result = response.choices[0].message.content.strip()
    print("result:")
    print(result)
    return result
````

````python
comment_list = pd.read_csv('comments_ì˜¥ì”¨ë¶€ì¸ì „_keyword.csv')
````

````python
comment_list
````

````python
acting_negative = comment_list[comment_list['ìºìŠ¤íŒ…'] == 'negative']
acting_negative = acting_negative.reset_index(drop=True)
acting_negative
````

````python
results = []
````

````python
results
````

````python
for i in range(len(acting_negative)):
  print(i)
  negative_reason = detail_info(acting_negative.loc[i, "comment"])
  negative_reason = negative_reason.strip("```json").strip("```").strip()
  negative_reason = json.loads(negative_reason)
  for reason in negative_reason:
        results.append({
            "comment": acting_negative.loc[i, "comment"],
            "actor": reason["actor"],
            "reason": reason["reason"]
        })
````

````python
df = pd.DataFrame(results)
df
````

````python
df.to_csv("ì˜¥ì”¨ë¶€ì¸ì „_casting_negative.csv", index=False)
````

````python
# ìœ ì‚¬ or ë™ì¼í•œ ì¹´í…Œê³ ë¦¬ëŠ” í•˜ë‚˜ë¡œ ë¬¶ê¸°
def clear_actor(col1, col2):
  global grouped_df
  plus_comment = grouped_df.loc[col1,"reason"]+grouped_df.loc[col2,"reason"]
  grouped_df.at[col1,"reason"] = plus_comment
  plus_count = grouped_df.loc[col1,"count"]+grouped_df.loc[col2,"count"]
  grouped_df.at[col1,"count"] = plus_count
````
