```python
import pandas as pd
from googleapiclient.discovery import build
```

```python
comment_list=pd.DataFrame(columns=["title", "comment"])
```

```python
# API 키
api_key = "my_youtube_api_key"

# YouTube API 클라이언트 생성
youtube = build("youtube", "v3", developerKey=api_key)
```

```python
# 대상 동영상 ID와 드라마 제목
video_id = "nfjzOsuBEbQ"
drama_title = "중증외상센터"
```

```python
# 댓글 가져오기 함수
def get_all_comments(video_id):
    comments = []  # 모든 댓글 저장 리스트
    next_page_token = None

    while True:
        # API 요청
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,              # 한 번에 가져올 댓글 수 (최대 100개)
            pageToken=next_page_token    # 다음 페이지 토큰
        ).execute()

        # 댓글 데이터 처리
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)

        # 다음 페이지 토큰 확인
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break  # 더 이상 페이지가 없으면 종료

    return comments
```

```python
# 댓글 데이터 가져오기
all_comments = get_all_comments(video_id)
```

```python
# 데이터프레임 생성
data = {"title": [drama_title] * len(all_comments), "comment": all_comments}
new_comment = pd.DataFrame(data)
comment_list = pd.concat([comment_list, new_comment], ignore_index=True)
```

```python
# 데이터프레임 출력
comment_list
```

```python
comment_list.to_csv("comments_중증외상센터-2.csv", index=False, encoding="utf-8-sig")
```
