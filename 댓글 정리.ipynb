```python
import html
import pandas as pd
import re
```

# 오징어 게임2

```python
comment = pd.read_csv('comments_오징어게임2_after_filtering.csv')
```

```python
for i in range(len(comment)):
  comment.loc[i,'comment'] = html.unescape(comment.loc[i,'comment'])
  comment.loc[i,'comment'] = re.sub(r"<br\s*/?>", "\n", comment.loc[i,'comment'])
  comment.loc[i,'comment'] = re.sub(r'\n+', '\n', comment.loc[i,'comment'])
```

```python
comment
```

```python
comment.to_csv('comments_오징어게임2_after_filtering.csv', index=False)
```
